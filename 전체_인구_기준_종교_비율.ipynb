```python
# 1. 나눔고딕 폰트 설치
!apt-get update -qq
!apt-get install -y fonts-nanum

# 2. matplotlib에 폰트 강제 적용 (폰트 캐시 재생성 포함)
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl
import os

# 새로 설치된 폰트 경로 확인
font_dirs = ['/usr/share/fonts/truetype/nanum']
font_files = fm.findSystemFonts(fontpaths=font_dirs)

# 폰트 등록
for font_file in font_files:
    fm.fontManager.addfont(font_file)

# 'NanumGothic'이 실제 등록되었는지 확인
mpl.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 엑셀 불러오기
df = pd.read_excel('종교별 인구 - 2015년 수정.xlsx', sheet_name='데이터')

# 전국 기준 데이터
no_religion = df.loc[0, '종교없음-계']
protestant = df.loc[0, '기독교(개신교)']
buddhism = df.loc[0, '불교']
catholic = df.loc[0, '기독교(천주교)']

# 라벨 및 비율
labels = ['무종교', '개신교', '불교', '천주교']
sizes = [no_religion, protestant, buddhism, catholic]

# 그래프 그리기
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('전체 인구 기준 무종교 / 개신교 / 불교 / 천주교 비율', pad=30)  # 여기 pad 추가!
plt.axis('equal')
plt.tight_layout()
plt.show()
```
